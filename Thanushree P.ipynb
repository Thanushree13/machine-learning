# 5.Percentage of Cloths bought by men and women

```python
# filter the dataset to include only purchases made by men and women separately
men = df[df["Gender"] == "M"]
women = df[df["Gender"] == "F"]
```

```python
# filter the dataset to include only clothing
clothing = df[(df["prod_cat"] == "Clothing")]
```

```python
total_clothing = len(clothing)
```

```python
total_clothing_purchases_by_men = len(men[men["prod_cat"].isin(["Clothing"])])
total_clothing_purchases_by_women = len(women[women["prod_cat"].isin(["Clothing"])])
```

```python
percentage_clothing_purchases_by_men = (total_clothing_purchases_by_men / total_clothing) * 100
```

```python
percentage_clothing_purchases_by_women = (total_clothing_purchases_by_women / total_clothing) * 100
```

```python
print(f"Percentage of clothing purchases made by men: {percentage_clothing_purchases_by_men:.2f}%")
```

```python
print(f"Percentage of clothing purchases made by women: {percentage_clothing_purchases_by_women:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Data
percentages = [percentage_clothing_purchases_by_men, percentage_clothing_purchases_by_women]
labels = ['Men', 'Women']

# Create bar graph
fig, ax = plt.subplots()
ax.bar(labels, percentages)
ax.set_title('Percentage of Clothing Purchases by Gender')
ax.set_ylabel('Percentage')
ax.set_ylim([0,100])
ax.set_yticks(range(0, 101, 10))
ax.set_xlabel('Gender')

# Show the plot
plt.show()
```

# 8. Type of Book  selling the most

```python
books = df[(df["prod_cat"] == "Books")]

# Group the DataFrame by book title and sum the sales
books_type= books.groupby('prod_subcat')['Qty'].sum()

# Sort the book sales in descending order and select the top 5
books_sold = books_type.sort_values(ascending=False)

# Print the most sold books and their sales
print(books_sold)
```

```python
import matplotlib.pyplot as plt

# Get the top 10 most sold book subcategories
top_books_sold = books_sold[:5]

# Create a bar chart to visualize the sales of top 10 book subcategories
plt.bar(top_books_sold.index, top_books_sold.values)

# Set the title and axis labels
plt.title("Top  Most Sold Book Subcategories")
plt.xlabel("Book Subcategory")
plt.ylabel("Number of Books Sold")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the chart
```

# 12. Who reads books more (Men/Women)

```python
Men = df[(df["Gender"] == "M") & (df["prod_cat"] == "Books")]
Bymen = Men["prod_cat_code"].sum()
print(Bymen)
```

```python
Women = df[(df["Gender"] == "F") & (df["prod_cat"] == "Books")]
ByWomen = Women["prod_cat_code"].sum()
print(ByWomen)
```

```python
if Bymen > ByWomen:
    print("Men read more books than women.")
elif Bymen < ByWomen:
    print("Women read more books than men.")
else:
    print("Men and Women read the same number of books.")
```

# 13. Type of books men read more

```python
# Filter the DataFrame to include only the rows where prod_cat is "Books"
books_df = df[df['prod_cat'] == 'Books']
#dataframe which shows only men
men_books_df = books_df[books_df['Gender'] == 'M']
# Group the data by gender and qty
grouped_data = men_books_df.groupby('prod_subcat')['Qty'].sum().reset_index(name='total_count')
#sorting into ascending 
sorted_data = grouped_data.sort_values('total_count', ascending=False)
# Get the types of books
most_popular_subcat = sorted_data.iloc[0]['prod_subcat']#integer location
print("The type of book that men read more is:", most_popular_subcat)
```
