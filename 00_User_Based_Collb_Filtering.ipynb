# <center>USER BASED COLLABORATIVE FILTERING</center>

#### There is a famous quote about the customer relationship. The summary of the quote is "Customers dont know what they want until we show them." So recommendation systems will help customers to find information, products and services they might not have thoght of. 

There are 2 steps involved in user based collaborative filtering.
1. Finding similar users

2. Recommendation phase: Predicting the rating of the missing item

We will consider a simple example with 4 ysers and predict the user 0 rating for I5. We assume that all other users {1,2,3} are neighbors to the user 0.

```python
#import sklearn library
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#dictionary with usernames their watched web series and ratings out of 5

dataset={'I1': [5,3,4,3], 'I2': [4,1,3,3], 'I3': [1,2,4,1], 'I4' : [4,3,3,5], 'I5' : [0,3,5,4] }
```

```python
#create a data frame of this dataset
dataset_df=pd.DataFrame(dataset)

dataset_df
```

```python
dataset_2 = dataset_df.drop('I5', axis = 1)
row_mean = dataset_2.mean(axis=1)
dataset_2 = dataset_2.sub(dataset_2.mean(axis=1), axis=0)
print(row_mean)
dataset_2
```

## Phase 1: Pairwise target user similarity with other users

```python
# custom function to implement cosine similarity between two users
def user_similarity(u1,u2):
    cs = cosine_similarity(u1,u2)
    return cs[0][0]
#print(type(user_similarity([dataset_2.iloc[0]],[dataset_2.iloc[1]])))
sim = []

for i in range(3):
    sim.append(user_similarity([dataset_2.iloc[0]],[dataset_2.iloc[i+1]]))
 #   print(user_similarity([dataset_2.iloc[0]],[dataset_2.iloc[i]]))
print('Similarity : ', sim)
```

```python
row_mean = list(row_mean)
print(type(row_mean))
```

# Phase 2: Recommendation Phase :Prediction of missing rate item

```python
uk_rating = 0;
for i in range(3):
    uk_rating += (dataset_df.loc[i+1].at['I5']-row_mean[i+1])*sim[i]
#print(type(uk_rating))
#print()
print('Predicted rating of user 1 for I5 is : ', row_mean[0]+ uk_rating/np.linalg.norm(sim, ord=1))
```
