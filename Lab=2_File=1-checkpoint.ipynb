```python
import math


def euclidean_dist(x_1, x_2):
  """Calculate the euclidean distance between two vector 
  Args: 
  X_1 & X_2: vectors of length n
  Returns : 
  A distance value ( scalar)
  """
  diff = [math.pow((x_1 - x_2), 2) for x_1, x_2 in zip(x_1, x_2)]
  return math.sqrt(sum(diff))

```

```python
a = [1.0, 3.2, 4.8, 0.1]
b = [0.1, 5.2, 1.9, 4.2]
print(euclidean_dist(a, b))
```
